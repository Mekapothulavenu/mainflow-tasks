```python
df.trending_date=df['trending_date'].apply(lambda x : datetime.datetime.strptime(x,'%y.%d.%m'))
df.head(3)
```

```python
df.publish_time	=pd.to_datetime(df["publish_time"])
df.head(3)
```

```python
import seaborn as sns
# Set the style for seaborn plots
sns.set(style="whitegrid")

# Task 1: Distribution of variables (views, likes, dislikes, and comment counts)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Distribution of Views
sns.histplot(df['views'], bins=50, kde=True, ax=axs[0, 0], color='blue')
axs[0, 0].set_title('Distribution of Views')

# Distribution of Likes
sns.histplot(df['likes'], bins=50, kde=True, ax=axs[0, 1], color='green')
axs[0, 1].set_title('Distribution of Likes')

# Distribution of Dislikes
sns.histplot(df['dislikes'], bins=50, kde=True, ax=axs[1, 0], color='red')
axs[1, 0].set_title('Distribution of Dislikes')

# Distribution of Comment Count
sns.histplot(df['comment_count'], bins=50, kde=True, ax=axs[1, 1], color='purple')
axs[1, 1].set_title('Distribution of Comment Count')

plt.tight_layout()
plt.show()
```

```python
# Task 2: Detect Outliers using Box Plots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Box Plot for Views
sns.boxplot(df['views'], ax=axs[0, 0], color='blue')
axs[0, 0].set_title('Box Plot of Views')

# Box Plot for Likes
sns.boxplot(df['likes'], ax=axs[0, 1], color='green')
axs[0, 1].set_title('Box Plot of Likes')

# Box Plot for Dislikes
sns.boxplot(df['dislikes'], ax=axs[1, 0], color='red')
axs[1, 0].set_title('Box Plot of Dislikes')

# Box Plot for Comment Count
sns.boxplot(df['comment_count'], ax=axs[1, 1], color='purple')
axs[1, 1].set_title('Box Plot of Comment Count')

plt.tight_layout()
plt.show()
```

```python
# Task 3: Correlation between numerical variables (views, likes, dislikes, comment count)
# Creating a correlation heatmap
plt.figure(figsize=(8,6))
corr_matrix = df[['views', 'likes', 'dislikes', 'comment_count']].corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
