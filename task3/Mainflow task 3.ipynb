```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("householdtask3.csv")
df 
```

```python
df.isnull().sum()
```

```python
plt.scatter(df['year'], df['own']) #Creating Scatter Plot with Year against own

plt.title('Scatter Plot') #Adding title to Scatter Plot

plt.xlabel('Year') #Setting Label name to x-axis

plt.ylabel('Own') #Setting label name to y-axis

plt.show()#displaying Scatter Plot
```

```python
plt.plot(df['year']) #Creating line chart using plot()

plt.plot(df['own']) #Creating Line chart using plot(

plt.title('Line Chart') #Adding title to Line Chart

plt.xlabel('Year') #Setting Label name to x-axis

plt.ylabel('Own') #Setting Label name to y-axis

plt.show()#displaying Line Chart
```

```python
plt.bar(df['year'], df['own']) #Creating Scatter Plot with Year against own

plt.title('Scatter Plot') #Adding title to Scatter Plot

plt.xlabel('Year') #Setting Label name to x-axis

plt.ylabel('Own') #Setting label name to y-axis

plt.show()#displaying Scatter Plot
```

```python
plt.hist(df['income'], bins=30, edgecolor='black', alpha=0.7) #Creating Histogram Plot 
plt.title("Histogram plot")#Adding title to Histogram Plot
plt.show()#displaying Histogram Plot
```
